```python
path = '/Users/lucia/phd_work/cognitive_distortion/'
def descriptiveStats(inputFile, outputFile):
    moodVec = pd.read_csv(inputFile)
    moodVec = moodVec.drop(['userid', 'Unnamed: 0'], axis = 1)
    stats = moodVec.describe()
    moodVec.describe()
    stats.to_csv(outputFile)
    return moodVec

inputFile = path + 'newScripts/moodVector/moodVectorsData/MoodVecAllVar.csv'
outputFile = path + 'newScripts/moodVector/moodVectorsData/reportTables/MoodVecStats.csv'
moodVec = descriptiveStats(inputFile, outputFile)
```

```python
def VarPlots(file):
    for column in file:
        emptyTran = sns.distplot(file[column])
        print(plt.show())
       
VarPlots(moodVec)
```
