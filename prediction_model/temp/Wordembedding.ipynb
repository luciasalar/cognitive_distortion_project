```python
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
import pandas as pd
import csv
import re
import gensim

```

```python
# Importing the dataset
data = pd.read_csv('../../data/self_label_distortion2.csv')
data.columns
```

```python
class MyFea:
    def __init__(self, text):
        self.text = text
        self.userid = hash(self.text)
        self.label = []
        self.vectors = []
        self.meanVec = []
        
    def __hash__(self):
        return self.quoteID
    
objects = {}
with open('../../data/self_label_distortion2.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row['text'])
        texthash = hash(row['text'])
        if texthash not in objects:
            objects[texthash] = MyFea(row['text'])
        objects[texthash].label.append(row['negative_yn_self'])

   
```

```python
count = 0 
for item in objects:
    print (objects[item].text)
    count = count + 1
    if count > 3:
        break
```

```python
def preprocess(sent):
    #remove punctustion\n",
    sent = re.sub(r'[^\w\s]','',sent)
    words = sent.split()
    new_words = []
    for w in words:     
        new_words.append(w.lower())
    return ' '.join(new_words)


def preProcessed(objects):
    doc = []
    for item in objects:
        processed = preprocess(objects[item].text)
        doc.append(processed.split())
    return doc

docs = preProcessed(objects)
```

```python
docs
```

```python
# build vocabulary and train model
#mymodel = gensim.models.Word2Vec(iter = 1)
#mymodel.build_vocab([docs])
mymodel = gensim.models.Word2Vec(
    docs,
    size=150,
    window=10,
    min_count=2,
    workers=10)
#mymodel.save("word2vec.model")
#mymodel = Word2Vec.load("word2vec.model")

mymodel.train(['messed'], total_examples=len(docs), epochs=10)
#mymodel.train([docs], total_examples=len(docs), epochs=10)


```
