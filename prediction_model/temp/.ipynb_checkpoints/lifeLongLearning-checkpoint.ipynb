```python
class MyFea:
    def __init__(self, text):
        self.text = text
        self.userid = hash(self.text)
        self.label = []
    def __hash__(self):
        return self.quoteID
    
objects = {}
with open('../../data/self_label_distortion2.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row['text'])
        texthash = hash(row['text'])
        if texthash not in objects:
            objects[texthash] = MyFea(row['text'])
        objects[texthash].label.append(row['negative_yn_self'])

def preprocess(sent):
    #remove punctustion\n",
    sent = re.sub(r'[^\w\s]','',sent)
    words = sent.split()
    new_words = []
    for w in words:     
        new_words.append(w.lower())
    return ' '.join(new_words)  
```

```python
#building knowledge base, tfidf word count 
positive = {}
negative = {}
for item in objects:
    line = preprocess(objects[item].text)
    #print(int(objects[item].label[0]) == 1)
    if int(objects[item].label[0]) == 1:
        for w in line.split():
            if w not in positive:
                positive[w] = 1
            else:
                positive[w] += 1 
    else:
        for w in line.split():
            if w not in negative:
                negative[w] = 1
            else:
                negative[w] += 1
#     #print (objects[item].text)
```
