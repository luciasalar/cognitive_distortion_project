```python
class MyFea:
    def __init__(self, text):
        self.text = text
        self.userid = hash(self.text)
        self.label = []
        
    def __hash__(self):
        return self.quoteID
    
objects = {}
with open('../data/self_label_distortion2.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row['text'])
        texthash = hash(row['text'])
        if texthash not in objects:
            objects[texthash] = MyFea(row['text'])
        objects[texthash].label.append(row['negative_yn_self'])

   
```

```python
def preProcessed(text):
    doc = []
    for item in objects:
        processed = gensim.utils.simple_preprocess(text)
        doc.append(processed)
    return doc

docs = preProcessed(objects[item].text)
```

```python
# build vocabulary and train model
mymodel = gensim.models.Word2Vec(
    docs,
    size=150,
    window=10,
    min_count=2,
    workers=10)

mymodel.train(docs, total_examples=len(docs), epochs=10)

```
